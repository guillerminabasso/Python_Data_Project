```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df_original = dataset['train'].to_pandas()
```

```python
df_original['job_posted_date'] = pd.to_datetime(df_original['job_posted_date'])
```

```python
median_salary_hour = df_original['salary_hour_avg'].median()
median_salary_year = df_original['salary_year_avg'].median()
```

```python
df_altered = df_original.copy()
print(f"ID of df_original: {id(df_original)}")
print(f"ID of df_altered: {id(df_altered)}")
```

```python
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_salary_hour)
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)
```

```python
df_unique = df_altered.copy()
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])
```

```python
print(f"ID of df_unique: {id(df_unique)}")
print(f"ID of df_altered: {id(df_altered)}")
```
