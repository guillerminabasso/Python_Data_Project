```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
```

```python
df = dataset['train'].to_pandas()
```

```python
df_job_country_salary = df.pivot_table(
    index='job_country',
    values='salary_year_avg',
    columns='job_title_short',
    aggfunc='mean')
```

```python
top_countries = df['job_country'].value_counts().head(6).index
```

```python
top_jobs = df['job_title_short'].value_counts().head(3).index
```

```python
df_job_country_salary = df_job_country_salary.loc[top_countries,top_jobs]
```

```python
df_job_country_salary.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Job Title and Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```
