```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
```

```python
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_salary = df[df['salary_year_avg'].notna()] # Way 1 to get df_salary
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])] # Another way for the same result
```

```python
import ast
```

```python
def skills_list (job_skills):
    if pd.notna(job_skills):
        return ast.literal_eval(job_skills) # Way 1 to execute the function to convert to a list
```

```python
df['job_skills'] = df['job_skills'].apply(
    lambda job_skills: ast.literal_eval(job_skills)
    if pd.notna(job_skills)
    else []) # Way 2 for doing the same but with lambda
```

```python
def projected_salary (row):
    if 'Senior' in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']
```

```python
df_salary['salary_year_inflated'] = df_salary.apply(projected_salary, axis=1)
```
