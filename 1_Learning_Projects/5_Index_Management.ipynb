```python
import pandas as ped
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
```

```python
df = dataset['train'].to_pandas()
```

```python
df.sample(5)
```

```python
df.index.name = 'job_index'
```

```python
df_belgium = df[df['job_country'] == 'Belgium']
```

```python
df_belgium.reset_index(inplace=True)
df_belgium
```

```python
df_belgium.set_index('job_index', inplace=True)
df_belgium
```

```python
df_belgium_jobs = df_belgium.pivot_table(
    index='job_title_short',
    values='salary_year_avg',
    aggfunc=['mean', 'min', 'max']
)
df_belgium_jobs.sort_values(by=('mean', 'salary_year_avg'), ascending=False)
```
