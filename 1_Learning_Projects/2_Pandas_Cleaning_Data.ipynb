```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[0:10, 'salary_rate':'salary_hour_avg']
```

```python
median_salary_year = df['salary_year_avg'].median()
median_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled.iloc[:10][['salary_year_avg', 'salary_hour_avg']]
```

```python
df_filled = df_filled.drop_duplicates()
print(f"Original number of rows: {len(df)}")
print(f"Number of rows after cleaning: {len(df_filled)}")
print(f"Number of rows removed: {len(df) - len(df_filled)}")
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])
print(f"Original number of rows: {len(df_filled)}")
print(f"Number of rows after cleaning: {len(df_unique)}")
print(f"Number of rows removed: {len(df_filled) - len(df_unique)}")
```
